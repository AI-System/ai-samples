# 2.对Lena图像，分解得到RGB分量及HSV分量，显示各分量，并对结果进行比较说明

```python
import cv2 as cv

filename = 'screenshot/lena.png'
img = cv.imread(filename)

# python的分离通道其他形式
'''
b = img[:, :, 0]
g = img[:, :, 1]
r = img[:, :, 2]
'''

# opencv 分离bgr通道
b, g, r = cv.split(img)

cv.imshow("Blue", b)
cv.imwrite('screenshot/lena_b.png', b)
cv.waitKey(0)

cv.imshow("Green", g)
cv.imwrite('screenshot/lena_g.png', g)
cv.waitKey(0)

cv.imshow("Red", r)
cv.imwrite('screenshot/lena_r.png', r)
cv.waitKey(0)

hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV) # 将bgr图像转化为hsv
h, s, v = cv.split(hsv) # 分离各个通道

# 显示hsv的各个通道
cv.imshow("Hue", h)
cv.imwrite('screenshot/lena_h.png', h)
cv.waitKey(0)

cv.imshow("Saturation", s)
cv.imwrite('screenshot/lena_s.png', s)
cv.waitKey(0)

cv.imshow("Value", v)
cv.imwrite('screenshot/lena_v.png', v)
cv.waitKey(0)

cv.destroyAllWindows()
```

其他备注：这里分量，我是按照通道来处理的，得到的图片都是单通道图片, 如果想要得到三通道的分量图片，比如蓝色三通道图片，蓝色通道保留，其他两个通道转化成0矩阵处理，如下代码：

```python
import cv2 as cv
import numpy as np

filename = 'screenshot/lena.png'
img = cv.imread(filename)

# 生成一个值为0的单通道数组，用于通道的合并
zeros = np.zeros(img.shape[:2], dtype = "uint8")

# 分离三个通道
b, g, r = cv.split(img)

# 使用merge生成一个三通道的纯色图片
bc = cv.merge([b, zeros, zeros])
cv.imshow("Blue Component", bc)
cv.imwrite('screenshot/lena_blue.png', bc)
cv.waitKey(0)

gc = cv.merge([zeros, g, zeros])
cv.imshow("Green Component", gc)
cv.imwrite('screenshot/lena_green.png', gc)
cv.waitKey(0)

rc = cv.merge([zeros, zeros, r])
cv.imshow("Red Component", rc)
cv.imwrite('screenshot/lena_red.png', rc)
cv.waitKey(0)
```
